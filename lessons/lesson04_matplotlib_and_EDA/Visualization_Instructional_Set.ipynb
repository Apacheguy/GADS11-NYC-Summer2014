###Ticks, Labels and Legends

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label='one') 
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')
ax.set_title('Three random lines')
ax.legend(loc='best')
plt.show()

#Exercise: Change the color and line style of each line using the minimum amount of characters
```

###Bar plots

```python
titanic = read_csv("data/titanic.csv")
```

```python
titanic
```

```python
# Append .plot to dataframe
titanic.groupby('pclass').survived.sum().plot(kind='bar')
```

```python
titanic.groupby(['sex','pclass']).survived.sum().plot(kind='barh')
```

```python
death_counts = crosstab([titanic.pclass, titanic.sex], titanic.survived.astype(bool))
death_counts.plot(kind='bar', stacked=True, color=['black','gold'], grid=False)
```

```python
# Try survival rate
death_counts.div(death_counts.sum(1).astype(float), axis=0).plot(kind='barh', stacked=True, color=['black','gold'])
```

###Histograms and Density Plots

```python
# Histogram of a particular column
titanic.fare.hist(grid=False,bins=25)
```

```python
# Density plot -- Estimate the 'true' distribution
titanic.fare.dropna().plot(kind='kde', xlim=(0,100))
```

```python
titanic.fare.hist(bins=25, normed=True, color='lightseagreen')
```

```python
titanic.fare.dropna().plot(kind='kde', xlim=(0,600), style='r--')
```

```python
# Combine pandas plots by running them together -- if pandas thinks they can be overlapped, it will
titanic.fare.hist(bins=25, normed=True, color='lightseagreen')
titanic.fare.dropna().plot(kind='kde', xlim=(0,600), style='r--')
titanic.boxplot(column='fare',by='pclass',grid=False)
```

```python
titanic.groupby('pclass').fare.sum().plot(kind='bar')
```

```python
# Boxplots view distributions from above
titanic.boxplot(column='fare',by='pclass',grid=False)

```

###Scatter Plots

```python
# Combine pandas with matplotlib for scatterplots
baseball = read_csv("data/baseball.csv")
baseball.head()
```

```python
plt.scatter(baseball.ab, baseball.h)
xlim(0, 700); ylim(0, 200)
```

```python
plt.scatter(baseball.ab, baseball.h, s=baseball.hr*10, alpha=0.5)
xlim(0, 700); ylim(0, 200)
```

```python
plt.scatter(baseball.ab, baseball.h, c=baseball.hr, s=40, cmap='hot')
xlim(0, 700); ylim(0, 200);
```

```python
_ = scatter_matrix(baseball.loc[:,'r':'sb'], figsize=(12,8), diagonal='kde')
```

```python
# Trellis Plots and GoG

#Create a RPlot object that merely relates two variables in the dataset
#Add a grid that will be used to condition the variables by both passenger class and sex
#Add the actual plot that will be used to visualize each comparison
#Draw the visualization
from pandas.tools.rplot import *

titanic = titanic[titanic.age.notnull() & titanic.fare.notnull()]

tp = RPlot(titanic, x='age')
tp.add(TrellisGrid(['pclass', 'sex']))
tp.add(GeomDensity())
_ = tp.render(gcf())
```

### Seaborn

```python
#With Titanic Data
import seaborn as sns
sns.set(style="darkgrid")

df = sns.load_dataset("titanic")

pal = dict(male="#6495ED", female="#F08080")
g = sns.lmplot("age", "survived", col="sex", hue="sex", data=df,
               palette=pal, y_jitter=.02, logistic=True)
g.set(xlim=(0, 80), ylim=(-.05, 1.05))
```

```python
# Seaborn with Anscombe Data
import seaborn as sns
sns.set(style="ticks")

df = sns.load_dataset("anscombe")
sns.lmplot("x", "y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", size=4,
           scatter_kws={"s": 50, "alpha": 1})
```
